## 1. Tri par insertion

**Compléter le code de la fonction suivante** pour qu'elle réalise le tri par insertion étudié en activité

```
def tri_par_insertion(tab):
    for i in range (1, len(tab) ) :
        j = i
        x = tab[i]
        while j > 0 and tab[j-1] > x : 
            tab[j] = tab[j-1]
            j = j-1
        
        tab[j] = x
    return tab
            
```

```
# test 1
tri_par_insertion([10,50,70,20,0,80,30])
```

```
# test 2
tri_par_insertion([50,50,70,50,0,80,0])
```

#### chrono
utiliser les fonctions déjà programmées dans ce notebook pour compléter la fonction suivante, qui mesure le temps d'exécution du tri par insertion sur un tableau aléatoire de taille n.

```
def chrono_tri_sur_taille(n):
    ''' renvoie le temps (en milliseconde) d'exécution du tri par selection
    sur un tableau aléatoire de 'n' nombres flottants'''
    tab = tableau_aleatoire(n)
    start = time()
    tri_par_insertion(tab)
    stop = time()
    delai = stop - start
    return delai * 1000
```

```
chrono_tri_sur_taille(1000)
```

```
chrono_tri_sur_taille(2000)
```

```
chrono_tri_sur_taille(3000)
```

Comparer le temps de tri pour un tableau aléatoire, et pour un tableau déjà trié. 

```
from random import random

def tableau_aleatoire(n):
    return [ 100*random() for i in range(n) ]

def tableau_trie(n):
    return [ i for i in range(n) ]
```

```
def chrono_deja_tri_sur_taille(n):
    ''' renvoie le temps (en milliseconde) d'exécution du tri par selection
    sur un tableau aléatoire de 'n' nombres flottants'''
    tab = tableau_trie(n)
    start = time()
    tri_par_insertion(tab)
    stop = time()
    delai = stop - start
    return delai * 1000
```

```
chrono_tri_sur_taille(10000) , chrono_deja_tri_sur_taille(10000)
```
