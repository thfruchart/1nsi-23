Un [fichier csv](https://raw.githubusercontent.com/thfruchart/1nsi/main/proj/nat2021.csv) contient, pour chaque année et pour chaque prénom recensé, le nombre de fois où ce prénom a été déclaré à l'état civil français.

Si un prénom a été donné moins de deux fois en une année, il n'est pas recensé, mais il est comptabilisé dans les "prénoms rares" de l'année.

Certaines années sont notées XXXX : cette notation regroupe toutes les années où un prénom a été considéré comme "prénom rare", et donne le nombre total sur l'ensemble de ces années.

### lecture du fichier csv
Le code ci-dessous permet d'importer le contenu du fichier dans une liste de dictionnaires.

```python
import csv

def lecture_fichier(nom_fichier):
    "lit un fichier csv et renvoie une liste de dictionnaires dont les clés sont les descripteurs"
    with open(nom_fichier, "r") as fichier:
        return [ligne for ligne in csv.DictReader(fichier, delimiter=',')]
```

```python
prenoms = lecture_fichier('nat2021.csv')
```
