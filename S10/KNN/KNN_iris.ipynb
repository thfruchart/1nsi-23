```
import csv
import matplotlib.pyplot as plt
from operator import itemgetter


def creerListes(fichier):
    #fichier est une chaine de caractères
    #assez clair et joli : on recup des listes Python faciles à traiter
        
    temp= open(fichier,newline='')
    L=list(csv.reader(temp))
                       
    liste_sepal=[float(L[i][2]) for i in range(1,len(L))]
    liste_petal=[float(L[i][4]) for i in range(1,len(L))]
    liste_variete=[L[i][5] for i in range(1,len(L))]
            
    return liste_sepal,liste_petal,liste_variete

def affichePoints():
    x,y,variete=creerListes("iris.csv")
    plt.plot(x[0:50],y[0:50],'ro',label='Setosa')
    plt.plot(x[50:100],y[50:100],'ob',label='Versicolor')
    plt.plot(x[100:],y[100:],'og',label='Virginica')
    plt.title("Fleurs d'iris")
    plt.grid()
    plt.xlabel('Largeur des sépales (cm)')
    plt.ylabel('Largeur des pétales (cm)')
    plt.legend()
    plt.show()
    


def distance(x1,y1,x2,y2):
    return ((x2-x1)**2+(y2-y1)**2)**0.5


def calcule_distances(x,y,liste_sepal,liste_petal,variete):
    tab=[]
    for i in range(len(variete)):
        tab.append([i+1,distance(x,y,liste_sepal[i],liste_petal[i]),variete[i]])
    return tab

def tri(tab):
    return sorted(tab,key=itemgetter(1))

def classe(k,tableauTrie):
    tab=tableauTrie[:k]
    dico={'Setosa':0,'Versicolor':0,'Virginica':0}
    for elm in tab:
        dico[elm[2]]+=1
    return dico

# définition des listes d'abscisses, ordonnées, et valeurs
x,y,v=creerListes("iris.csv")

affichePoints()
```

```
fleur1=(3.5,0.2)
fleur2=(3.0,1.4)
fleur3=(3.0,1.8)

k = 12
```

### Fleur 1

```
fleur = fleur1
```

#### RECHERCHE des K plus proches voisins de la fleur

```
tab=tri(calcule_distances(fleur[0],fleur[1],x,y,v))
print(classe(k,tab))
```

#### Ajout de la fleur sur le graphique en faisant apparaître les k plus proches voisins

```
plt.plot(fleur[0],fleur[1],'c^')
circle1 = plt.Circle(fleur, tab[k-1][1], color='c', fill=False)
plt.gca().add_patch(circle1)
plt.show()
```
