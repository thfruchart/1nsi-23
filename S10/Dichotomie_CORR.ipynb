## Exercice 1

On donne la fonction suivante : 

````
def recherche_dicho(t, v):
    g = 0 # indice gauche de la zone de recherche
    d = len(t) - 1  # indice droit de la zone de recherche
    while (g <= d) :
        m = (g + d) // 2  # indice médian de la zone de recherche
        if t[m] == v:
                return True
        elif t[m] < v:
                g = m + 1
        else:
                d = m - 1
    return False

def compte_recherche_dicho(t, v):
    cpt = 0
    g = 0 # indice gauche de la zone de recherche
    d = len(t) - 1  # indice droit de la zone de recherche
    while (g <= d) :
        cpt += 1
        m = (g + d) // 2  # indice médian de la zone de recherche
        if t[m] == v:
                return cpt
        elif t[m] < v:
                g = m + 1
        else:
                d = m - 1
    return cpt
````

#### Combien de fois la boucle while de la fonction `recherche_dicho`  sera-t-elle exécutée dans chacun des cas suivants ?

````
def recherche_incorrecte(t, v):
    g = 0 # indice gauche de la zone de recherche
    d = len(t) - 1  # indice droit de la zone de recherche
    while (g <= d) :
        m = (g + d) // 2  # indice médian de la zone de recherche
        if t[m] == v:
                return True
        elif t[m] < v:
                g = m + 2
        else:
                d = m - 2
    return False
````

Si g = m+2 est exécuté, la valeur 40 n'est plus dans la zone de recherche.

Si d = m-2 est exécuté, la valeur 20 n'est pas plus dans la zone de recherche.

## Exercice 3

````
def recherche_presque_correcte(t, v):
    g = 0 # indice gauche de la zone de recherche
    d = len(t) - 1  # indice droit de la zone de recherche
    while (g <= d) :
        m = (g + d) // 2  # indice médian de la zone de recherche
        print(f"g={g}, d={d} et m={m}")
        if t[m] == v:
                return True
        elif t[m] < v:
                g = m 
        else:
                d = m 
    return False
````

Trouver un tableau `t` **trié** et une valeur `v` pour lequels la fonction ci-dessus ne renvoie pas la valeur attendue.

On pourra choisir une valeur v qui ne figure pas dans le tableau t, et pour laquelle `recherche_incorrecte(t, v)` ... ne renvoie rien !

Lorsque l'instruction g = m est exécutée, cela ne permet pas toujours à g d'augmenter, car il arrive que g et m aient la même valeur.

Dans ce cas, la boucle while peut être infinie, et l'algorithme ne s'arrête jamais

Exemple : 

```
recherche_presque_correcte([10,20,30,40,50],50)
g=0, d=4 et m=2
g=2, d=4 et m=3
g=3, d=4 et m=3
g=3, d=4 et m=3
g=3, d=4 et m=3
```

````
def recherche_presque_correcte_avec_borne(t, v):
    g = 0 # indice gauche de la zone de recherche
    d = len(t) - 1  # indice droit de la zone de recherche
    cpt = 12
    while (g <= d) and cpt > 0 :
        cpt -= 1
        m = (g + d) // 2  # indice médian de la zone de recherche
        print(f"g={g}, d={d} et m={m}")
        if t[m] == v:
                return True
        elif t[m] < v:
                g = m 
        else:
                d = m 
    return False
````
