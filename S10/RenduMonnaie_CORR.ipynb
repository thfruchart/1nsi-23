# Question 1

Ecrire une fonction rendu_monnaie qui :  
* prend en arguments : 
  * la somme à rendre `somme`
  * la liste des valeurs des pièces `systeme` 
* et qui renvoie la liste des effectifs des pièces à rendre, pour chacune des valeurs du système, avec un algorithme glouton.

Par exemple :  
```
>>> rendu_monnaie(93, [50,20,10,5,2,1])  
[1,2,0,0,1,1]
```

````
def rendu_monnaie(somme, systeme):
    resu = []
    for piece in systeme :
        nb_pieces_rendues = somme // piece
        resu.append(nb_pieces_rendues)
        somme = somme % piece
        #somme -= piece * nb_pieces_rendues
        print(piece, nb_pieces_rendues, somme)
        
    return resu
    
````

````
rendu_monnaie(93, [50,20,10,5,2,1])  
````

### Ecrire une nouvelle version de la fonction rendu_monnaie, qui :
* prend en arguments : 
  * la somme à rendre `somme`
  * la liste des valeurs des pièces `systeme` 
* et qui renvoie la liste des pièces à rendre, la plus courte possible.

Par exemple :  
```
>>> rendu_monnaie(93, [50,20,10,5,2,1])  
[50,20,20,2,1]
```



````
def rendu_monnaie(somme, systeme):
    resu = []
    for piece in systeme :
        nb_pieces_rendues = somme // piece
        somme = somme % piece
        print(piece, nb_pieces_rendues, somme)
        for _ in range(nb_pieces_rendues):
            resu.append(piece)
        
    return resu
````

````
def rendu_monnaie(somme, systeme):
    resu = []
    for piece in systeme :
        nb_pieces_rendues = somme // piece
        somme = somme % piece
        resu = resu + [piece]*nb_pieces_rendues
        
    return resu
````
