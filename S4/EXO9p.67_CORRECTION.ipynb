```python
table1 = [('id_exp',  'T' , 'tps'),
          (    1   , 20.2 ,  56  ),
          (    2   , 15.5 ,  85  ),
          (    3   , 18.6 , 120  )]
```

```python
table2 = [('id_exp',  'T' , 'tps'),
          (    1   , 53.6 , 210  ),
          (    2   , 62.6 ,  81  ),
          (    3   , 66.2 ,  70  )]
```

```python
def fusion(t1,t2):
    """fusionne t1 et t2
    en renumérottant les expériences de t2
    et en convertissant les températures de t2 en degrés Celsius"""
    table2_conv = [] # liste vide 
    for ligne in t2[1:] : # pour chaque ligne (sauf descipteurs) de t2
        id, T, tps = ligne
        id += len(t1) - 1
        T = (T-32) / 1.8 
        valeurs_converties =  (id, T, tps)
        table2_conv.append(valeurs_converties)
        
    table = table1 + table2_conv
    
    return table
```

```python
fusion(table1, table2)
```
