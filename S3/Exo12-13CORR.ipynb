## exercice  13 p. 53

écrire une fonction `ajoute_fruit(panier, fruit, quantite)` qui ajoute le fruit dans le panier, avec la quantité indiquée.

Tenir compte des deux cas : 
- fruit déjà présent dans le panier
- nouveau fruit à ajouter

La fonction renvoie le nouveau panier complété

```python
def ajoute_fruit(panier, fruit, quantite):
    if fruit in panier :
        panier[fruit] += quantite
    else :
        panier[fruit] = quantite
    return panier
```

```python
x = {'pomme': 2,'orange' : 2}
```

```python
ajoute_fruit(x, 'banane', 3)
```

```python
ajoute_fruit(x, 'orange', 3)
```
