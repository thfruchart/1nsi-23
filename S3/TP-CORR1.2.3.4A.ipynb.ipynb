
````python
# t = [i for i in range (1,11)]
s = 0
for i in range( len(t) ):
    v = t[i]
    s = s + v

# print('la somme des valeurs du tableau est : ')
# print( s )
````

````python
def total(t):
    s = 0
    for i in range( len(t) ):
        v = t[i]
        s = s + v
    return s
````

````python
total([10,20,30])
````
