# Types construits : PARCOURS avec une boucle `for`

Les types construits permettent de stocker **plusieurs valeurs** dans une seule variable.

Suivant le type choisi, il existe plusieurs manières de **parcourir** les valeurs contenues dans cette variable.
* avec les indices : type tuple ou list
* avec les clés : type dictionnaire
* ou en accédant directement aux valeurs elle-mêmes ! 

## Parcours d'un tableau 
### 1) avec les indices : `for i in range(len(t))`

Dans le code suivant :
*  la variable `i` parcourt les indices du tuple `t1`
* la variable `v` parcourt les valeurs  du tuple `t1`

```python
t1 = ('A', 'B', 'C', 'D')
for i in range(len(t1)) : 
    v = t1[i]
    print("indice :", i, "valeur : ", v)
```

Dans le code suivant :
*  la variable `i` parcourt les indices du tableau `t2`
* la variable `v` parcourt les valeurs  du tableau `t2`

```python
t2 = [10,20,30,40,50]
for i in range(len(t2)) : 
    v = t2[i]
    print("indice :", i, "valeur : ", v)
```

## Parcours d'un tableau 
### 2) en accédant directement aux valeurs d'une liste

Dans le code suivant 
*  la variable `v` parcourt les valeurs du tableau `t` sans utiliser les indices

```python
t = [10,20,30,40,50]
for v in t : 
    print("valeur : ", v)
```

## Parcours d'un dictionnaire
### 1) avec les clés du dictionnaire : `for k in d`

Dans le code suivant :
*  la variable `k` parcourt les clés du dictionnaire `d1`
* la variable `v` parcourt les valeurs  du dictionnaire `d1`

```python
d1 = {"A":'Alpha', "B": 'Bravo', "C":'Charlie'}

for k in d1:
    v = d1[k]
    print("clé :", k, "=> valeur : ", v)
```

remarque : pour parcourir les clés d'un dictionnaire, on peut aussi écrire `for k in d.keys():`

## Parcours d'un dictionnaire
### 2)  en accédant directement aux valeurs d'un dictionnaire

Dans le code suivant on utilise la méthode **`values()`**
* la variable `v` parcourt les valeurs  du dictionnaire `d2` (sans utiliser de clé)

```python
d2 = {"A":'Alpha', "B": 'Bravo', "C":'Charlie', "D" : 'Delta'}

for v in d2.values():
    print("valeur : ", v)
```

#### 2 bis)
* on peut aussi parcourir "directement" les clés **et** les valeurs avec la méthode **`items()`**

```python
grec = {"A":'Alpha', "B": 'Bêta', "G":'Gamma'}

for k, v in grec.items():
    print("clé :", k, "valeur : ", v)
```
